```python
value_heroin_lastyear_list = []
value_heroin_list = []
value_opioid_lastyear_list = []
value_opioid_list = []

for i in range(len(trans_data)):
    year = trans_data['year'][i]
    geoid = trans_data['GEO.id2'][i]
    if (geoid in report_heroin_raw['FIPS_Combined'].values):
        if (year in report_heroin_raw.loc[report_heroin_raw['FIPS_Combined'] ==geoid,'YYYY'].values):
            value_heroin_list.append(report_heroin_raw.loc[(report_heroin_raw['FIPS_Combined'] == geoid)&(report_heroin_raw['YYYY'] == year), 'DrugReports'].values[0])
        else: value_heroin_list.append(0)
        if (year-1 in report_heroin_raw.loc[report_heroin_raw['FIPS_Combined'] ==geoid,'YYYY'].values):
            value_heroin_lastyear_list.append(report_heroin_raw.loc[(report_heroin_raw['FIPS_Combined'] == geoid)&(report_heroin_raw['YYYY'] == year-1), 'DrugReports'].values[0])
        else: value_heroin_lastyear_list.append(0)
    else:
        value_heroin_list.append(0)
        value_heroin_lastyear_list.append(0)

    if (geoid in report_opioid_raw['FIPS_Combined'].values):
        if (year in report_opioid_raw.loc[report_opioid_raw['FIPS_Combined'] ==geoid,'YYYY'].values):
            value_opioid_list.append(report_opioid_raw.loc[(report_opioid_raw['FIPS_Combined'] == geoid)&(report_opioid_raw['YYYY'] == year), 'DrugReports'].values[0])
        else: value_opioid_list.append(0)
        if (year-1 in report_opioid_raw.loc[report_opioid_raw['FIPS_Combined'] ==geoid,'YYYY'].values):
            value_opioid_lastyear_list.append(report_opioid_raw.loc[(report_opioid_raw['FIPS_Combined'] == geoid)&(report_opioid_raw['YYYY'] == year-1), 'DrugReports'].values[0])
        else: value_opioid_lastyear_list.append(0)
    else:
        value_opioid_list.append(0)
        value_opioid_lastyear_list.append(0)
```

```python
trans_data['heroin_lastyear'] =value_heroin_lastyear_list
trans_data['opioid_lastyear'] =value_opioid_lastyear_list
trans_data['heroin'] = value_heroin_list
trans_data['opioid'] = value_opioid_list
```

```python
trans_data
```

```python
#将2010年去除掉
trans_data_for_train = trans_data.loc[trans_data['year'] !=2010,:]
trans_data_for_train.reset_index(drop =True, inplace=True)
```

```python
trans_data_for_train.head()
```

```python
trans_data_for_train.to_csv('trans_data_combine_history.csv', index =False)
```

```python
from sklearn.model_selection import train_test_split
#Train_Test_Split得到训练集和测试集
trainDF_all, testDF_all  = train_test_split(trans_data_for_train, test_size = 0.1, random_state = 89)
```

```python
x_train = trainDF_all.drop(['GEO.id2','year','heroin','opioid'],axis=1).values
x_test = testDF_all.drop(['GEO.id2','year','heroin','opioid'],axis=1).values
y_train_opioid = trainDF_all['opioid'].values
y_test_opioid = testDF_all['opioid'].values
y_train_heroin = trainDF_all['heroin'].values
y_test_heroin = testDF_all['heroin'].values
```

```python
def MAPE_List(y_true, y_pred): 
    return (y_true - y_pred) / y_true * 100
```

```python
from sklearn.metrics import mean_squared_error
from xgboost.sklearn import XGBRegressor
from math import sqrt
#xgb_opioid
xgb_opioid = XGBRegressor(learning_rate =0.01, n_estimators=1000, max_depth=8,
    min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
    nthread=4, scale_pos_weight=1, seed=27)
xgb_opioid.fit(x_train, y_train_opioid)
opioid_pred_xgb = xgb_opioid.predict(x_test)
rmse_xgb_opioid = sqrt(mean_squared_error(y_test_opioid, opioid_pred_xgb))
print('Test RMSE xgb_opioid: %.3f' %rmse_xgb_opioid)

#xgb_heroin
xgb_heroin = XGBRegressor(learning_rate =0.01, n_estimators=1000, max_depth=8,
    min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
    nthread=4, scale_pos_weight=1, seed=27)
xgb_heroin.fit(x_train, y_train_heroin)
heroin_pred_xgb = xgb_heroin.predict(x_test)
rmse_xgb_heroin = sqrt(mean_squared_error(y_test_heroin, heroin_pred_xgb))
print('Test RMSE xgb_heroin: %.3f' %rmse_xgb_heroin)
```

```python
import lightgbm as lgb
#对opioid训练LightGBM
# train
gbm_opioid = lgb.LGBMRegressor(objective='regression',
                        num_leaves=63,
                        learning_rate=0.01,
                        n_estimators=200)
gbm_opioid.fit(x_train, y_train_opioid)
# predict
y_pred_opioid = gbm_opioid.predict(x_test)
# eval
rmse_gbm_opioid = sqrt(mean_squared_error(y_test_opioid, y_pred_opioid))

print('Test RMSE lgb: %.3f' %rmse_gbm_opioid)
#mape_gbm = MAPE(y_test, y_pred)
#print('Test MAPE xgb: %.3f' %mape_gbm)

#对heroin训练LightGBM
# train
gbm_heroin = lgb.LGBMRegressor(objective='regression',
                        num_leaves=63,
                        learning_rate=0.01,
                        n_estimators=200)
gbm_heroin.fit(x_train, y_train_heroin)
# predict
y_pred_heroin = gbm_heroin.predict(x_test)
# eval
rmse_gbm_heroin = sqrt(mean_squared_error(y_test_heroin, y_pred_heroin))

print('Test RMSE lgb: %.3f' %rmse_gbm_heroin)
#mape_gbm = MAPE(y_test, y_pred)
#print('Test MAPE xgb: %.3f' %mape_gbm)
```
